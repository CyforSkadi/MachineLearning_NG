# 3-寻找最优参数
## 数据集：ex6data3.mat

```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
from sklearn.svm import SVC
```

```python
mat = sio.loadmat('ex6data3.mat')
X,y = mat['X'],mat['y']
Xval,yval = mat['Xval'],mat['yval']
```

```python
def plot_data():
    plt.scatter(X[:,0],X[:,1],c=y.flatten(),cmap='jet')
    plt.xlabel('x1')
    plt.ylabel('x2')
```

```python
plot_data()
```

```python
Cvalues = [0.01,0.03,0.1,0.3,1,3,10,30,100] #9
gammas = [0.01,0.03,0.1,0.3,1,3,10,30,100] #9
```

```python
best_score = 0
best_params = (0,0)
```

```python
for c in Cvalues:
    for gamma in gammas:
        svc = SVC(C=c,kernel='rbf',gamma=gamma)
        svc.fit(X,y.flatten())
        score = svc.score(Xval,yval.flatten())
        
        if score > best_score:
            best_score = score
            best_params = (c,gamma)
            
print(best_score,best_params)
```

```python
svc_best = SVC(C=0.3,kernel='rbf',gamma=100)
svc_best.fit(X,y.flatten())
```

```python
def plot_boundary(model):
    """绘制决策边界"""
    x_min,x_max = -0.6,0.4
    y_min,y_max = -0.7,0.6
    xx,yy = np.meshgrid(np.linspace(x_min,x_max,500),
                       np.linspace(y_min,y_max,500))
    z = model.predict(np.c_[xx.flatten(),yy.flatten()])
    zz = z.reshape(xx.shape)
    plt.contour(xx,yy,zz)
```

```python
plot_boundary(svc_best)
plot_data()
```
