# 2-线性不可分支持向量机
## 案例：使用高斯核函数解决线性不可分问题，并观察gamma值对模型复杂度的影响
### 数据集：ex6data2.mat

```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
from sklearn.svm import SVC
```

```python
data = sio.loadmat('ex6data2.mat')
data.keys()
```

```python
X,y = data['X'],data['y']
X.shape,y.shape
```

```python
def plot_data():
    plt.scatter(X[:,0],X[:,1],c=y.flatten(),cmap='jet')
    plt.xlabel('x1')
    plt.ylabel('x2')
```

```python
plot_data()
```

```python
svc1 = SVC(C=1,kernel='rbf',gamma=1)
svc1.fit(X,y.flatten())
```

```python
svc1.score(X,y.flatten())
```

```python
def plot_boundary(model):
    """绘制决策边界"""
    x_min,x_max = 0,1
    y_min,y_max = 0.4,1
    xx,yy = np.meshgrid(np.linspace(x_min,x_max,500),
                       np.linspace(y_min,y_max,500))
    z = model.predict(np.c_[xx.flatten(),yy.flatten()])
    zz = z.reshape(xx.shape)
    plt.contour(xx,yy,zz)
```

```python
plot_boundary(svc1)
plot_data()
```

### gamma=1分类效果

```python
svc50 = SVC(C=1,kernel='rbf',gamma=50)
svc50.fit(X,y.flatten())
svc50.score(X,y.flatten())
```

```python
plot_boundary(svc50)
plot_data()
```
