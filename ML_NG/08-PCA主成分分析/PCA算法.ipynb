# 1-PCA算法
## 案例：使用PCA进行二维数据降维
### 数据集：ex7data1.mat

```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
```

```python
mat = sio.loadmat('ex7data1.mat')
X = mat['X']
X.shape
```

```python
plt.scatter(X[:,0],X[:,1])
plt.show()
```

### 1.对X取均值化

```python
X_demean = X - np.mean(X,axis=0)
plt.scatter(X_demean[:,0],X_demean[:,1])
plt.show()
```

### 2.计算协方差矩阵

```python
C = X_demean.T @ X_demean / len(X_demean)
print(C)
```

### 3.计算特征值和特征向量

```python
U,S,V = np.linalg.svd(C)
U1 = U[:,0]
```

### 4.实现降维

```python
X_reduction = X_demean @ U1
```

```python
plt.scatter(X_demean[:,0],X_demean[:,1])
plt.plot([0,U1[0]],[0,U1[1]],c='r')
plt.show()
```

### 5.还原数据

```python
X_restore = X_reduction.reshape(50,1) @ U1.reshape(1,2) + np.mean(X,axis=0) #维度变换
```

```python
plt.scatter(X[:,0],X[:,1])
plt.scatter(X_restore[:,0],X_restore[:,1])
plt.show()
```
